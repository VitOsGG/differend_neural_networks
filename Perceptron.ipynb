**Персептрон**

Возможность классификации двух образов. Кошка\Собака.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5 #5 образов для обоих классов

# формируем случйные точки, которые необходимо будет классифицировать
x1 = np.random.random(N)
x2 = x1 + [np.random.randint(10)/10 for i in range(N)]
C1 = [x1, x2] #10 значений -5 пар (5 точек - 2 координаты на точку)

# формируем случйные точки, которые необходимо будет классифицировать
x1 = np.random.random(N)
x2 = x1 - [np.random.randint(10)/10 for i in range(N)] - 0.1 #-0.1 необходимо, чтобы точки точно были ниже прямой, относительной которой будет определяться классификация
C2 = [x1, x2]

f = [0, 1] #прямая под 45 граудсов по которой будет разделение (гиперплоскость)

w = np.array([-0.5, 0.5]) #Веса (позволяются правильно преобразовывать входные значения)
for i in range(N):
    x = np.array([C1[0][i], C1[1][i]])
    y = np.dot(w, x) #функция активации
    if y >= 0:
        print("Класс C1")
    else:
        print("Класc C2")
        
plt.scatter(C1[0][:], C1[1][:], s=10, c='red')
plt.scatter(C2[0][:], C2[1][:], s=10, c='blue')
plt.plot(f) #разделяющая прямая
plt.grid(True)
plt.show
```

```python
N = 5 #5 образов для обоих классов
b = 3 #коэффициент добавляющий смещения прямой по координатам

# формируем случйные точки, которые необходимо будет классифицировать
x1 = np.random.random(N)
x2 = x1 + [np.random.randint(10)/10 for i in range(N)] + b
C1 = [x1, x2]

# формируем случйные точки, которые необходимо будет классифицировать
x1 = np.random.random(N)
x2 = x1 - [np.random.randint(10)/10 for i in range(N)] - 0.1 + b #-0.1 необходимо, чтобы точки точно были ниже прямой, относительной которой будет определяться классификация
C2 = [x1, x2]

f = [0+b, 1+b] #прямая под 45 граудсов по которой будет разделение (гиперплоскость)

w2 = 0.5
w3 = -b*w2
w = np.array([-w2, w2, w3]) #Веса (позволяются правильно преобразовывать входные значения)
for i in range(N):
    x = np.array([C1[0][i], C1[1][i], 1])
    y = np.dot(w, x) #функция активации
    if y >= 0:
        print("Класс C1")
    else:
        print("Класc C2")
        
plt.scatter(C1[0][:], C1[1][:], s=10, c='red')
plt.scatter(C2[0][:], C2[1][:], s=10, c='blue')
plt.plot(f)
plt.grid(True)
plt.show
```

Задача XOR

```python
def act(x):
    return 0 if x <= 0 else 1 #активационная функция

def go(C):
    x = np.array([C[0], C[1], 1]) #входные значения, либо 0 либо 1. И последняя 1 - это bias (пороговое значение для смещения прямой)
    w1 = [1, 1, -1.5] #веса для 1 нейрона
    w2 = [1, 1, -0.5] #веса для 2 нейрона
    w_hidden = np.array([w1, w2]) #объеденяем веса в матрицу
    w_out = np.array([-1, 1, -0.5]) #веса для выходного нейрона
    
    sum = np.dot(w_hidden, x) #сумма накаждом нейроне
    out = [act(x) for x in sum] #пропускаем сумму через функцию активации
    out.append(1) #добавляем 1 (bias)
    out = np.array(out) #все превращаем в вектор
    
    sum = np.dot(w_out, out) #вектор умножаем на весовые коэффициенты для выходного нейрона(получаем сумму)
    y = act(sum) #сумму пропускаем через функцию активации, получаем выхходное значение нейронной сети
    return y

C1 = [(1,0), (0,1)]
C2 = [(0,0), (1,1)]

print( go(C1[0]), go(C1[1]) ) #Определен как 1
print( go(C2[0]), go(C2[1]) ) #Определен как 2
```
